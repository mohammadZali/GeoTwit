```python


import pandas as pd
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="django",
    password="django-user-password",
    database="citizen_base"
)

# Create a cursor object
cursor = conn.cursor()

# Execute SQL query
query = "SELECT * FROM boards_post;"
cursor.execute(query)

# Fetch all rows of data
result = cursor.fetchall()

# Convert the result into a Pandas DataFrame
df = pd.DataFrame(result, columns=[column[0] for column in cursor.description])

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
import spacy

df['Adress'] = 0
# Load SpaCy's English model
nlp = spacy.load('en_core_web_sm')

# Define the input text
for i in range(df.shape[0]):
    text = df['message'][i]
    doc = nlp(text)
    for entity in doc.ents:
    # Check if the entity is a location
        if entity.label_ == 'LOC' or entity.label_ == 'GPE':
            df['Adress'][i]= entity.text

df
```

```python
from geopy.geocoders import Nominatim
import pandas as pd

# Connect to database and fetch data into a pandas dataframe
# For example:
# df = pd.read_sql_query('SELECT * FROM my_table', con=my_db_connection)

# Initialize geocoder
geolocator = Nominatim(user_agent="my_application")

# Define function to geocode addresses
def geocode(address):
    location = geolocator.geocode(address)
    if location is not None:
        return location.latitude, location.longitude
    else:
        return None, None

# Apply geocoding function to address column in dataframe
df['latitude'], df['longitude'] = zip(*df['Adress'].apply(geocode))
```

```python
df = df[['message','Adress','latitude','longitude']]
df
```

```python
import folium
map = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=10)

# Add markers for each location with a popup message containing the message text
for index, row in df.iterrows():
    popup_html = f'<div style="display: flex; flex-direction: row; align-items: center;"><p style="margin-right: 10px;">{row["message"]}</p></div>'
    popup_text = row['message']
    popup = folium.Popup(popup_html, max_width=100000, max_height=1000)
    marker = folium.Marker([row['latitude'], row['longitude']], popup=popup_text)
    marker.add_to(map)

# Display the map
map.save('my_map.html')
```
